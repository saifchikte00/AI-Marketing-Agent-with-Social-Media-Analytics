```python
import pandas as pd
import numpy as np
import sqlite3
import hashlib
import datetime
import matplotlib.pyplot as plt
from textblob import TextBlob

print("✅ All Libraries Loaded Successfully")
```

```python
DB_NAME = "social_media_users.db"

conn = sqlite3.connect(DB_NAME)
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS users (
    user_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    email TEXT UNIQUE NOT NULL,
    password TEXT NOT NULL,
    created_at TEXT
)
""")

conn.commit()
conn.close()

print("✅ User Database Created Successfully")
```

```python
def hash_password(password):
    return hashlib.sha256(password.encode()).hexdigest()

print("✅ Password Security Function Ready")
```

```python
1: User Sign Up
```

```python
def user_signup():

    name = input("Enter Name: ")
    email = input("Enter Email: ")
    password = input("Enter Password: ")

    conn = sqlite3.connect(DB_NAME)
    cursor = conn.cursor()

    try:
        cursor.execute("""
        INSERT INTO users(name, email, password, created_at)
        VALUES (?, ?, ?, ?)
        """,
        (
            Name,
            Email,
            hash_password(Password),
            datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        ))

        conn.commit()

        print("\n✅ Sign Up Successful!")
        print("Welcome,", name)

    except sqlite3.IntegrityError:
        print("\n❌ Email already exists!")

    conn.close()
```

```python
user_signup()
```

```python
2: User Login
```

```python
current_user = None

def user_login():
    global current_user

    email = input("Enter Email: ")
    password = input("Enter Password: ")

    conn = sqlite3.connect(DB_NAME)
    cursor = conn.cursor()

    cursor.execute("""
    SELECT user_id, name, email, created_at
    FROM users
    WHERE email = ? AND password = ?
    """, (email, hash_password(password)))

    user = cursor.fetchone()
    conn.close()

    if user:
        current_user = {
            "user_id": user[0],
            "Name": user[1],
            "E
        Email": user[2],
            "created_at": user[3]
        }

        print("\n✅ Login Successful!")
        print("Welcome back,", current_user["name"])
    else:
        print("\n❌ Invalid Email or Password!")
```

```python
9: Sentiment Analysis
```

```python
def sentiment_analysis():

    text = input("Enter customer comment/review: ")

    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity

    if polarity > 0:
        sentiment = "Positive 😊"
    elif polarity < 0:
        sentiment = "Negative 😞"
    else:
        sentiment = "Neutral 😐"

    print("\n✅ Sentiment Analysis Result")
    print("="*50)
    print("Comment:", text)
    print("Polarity Score:", polarity)
    print("Sentiment:", sentiment)
```

```python
sentiment_analysis()
```

```python
11: Competitor Analysis
```

```python
def competitor_analysis():

    df = pd.read_csv("../data/social_media_data.csv")

    df["Engagement"] = df["Likes"] + df["Comments"] + df["Shares"]

    platform_engagement = df.groupby("Platform")["Engagement"].mean()

    print("\n✅ Competitor Analysis")
    print("="*60)

    print(platform_engagement)

    best_platform = platform_engagement.idxmax()

    print("\n🏆 Best Performing Platform:")
    print(best_platform)

    print("\n📈 Highest Average Engagement:")
    print(platform_engagement.max())
```

```python
def social_media_analytics_dashboard():

    df = pd.read_csv("../data/social_media_data.csv")

    df["Engagement"] = df["Likes"] + df["Comments"] + df["Shares"]

    print("\n✅ Social Media Analytics Dashboard")
    print("="*60)

    print(df[["Platform", "Post", "Likes", "Comments", "Shares", "Engagement"]])

    plt.figure(figsize=(10, 5))
    plt.bar(df["Post"], df["Engagement"])
    plt.xlabel("Posts")
    plt.ylabel("Engagement")
    plt.title("Post-wise Engagement Analysis")
    plt.xticks(rotation=45, ha="right")
    plt.show()

    plt.figure(figsize=(8, 5))
    platform_data = df.groupby("Platform")["Engagement"].sum()
    plt.bar(platform_data.index, platform_data.values)
    plt.xlabel("Platform")
    plt.ylabel("Total Engagement")
    plt.title("Platform-wise Engagement Analysis")
    plt.show()
```
